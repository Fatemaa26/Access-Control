```python
import pytesseract
from PIL import Image


path = '/content/drive/MyDrive/fatoom.png'
img = Image.open(path)

extracted = pytesseract.image_to_string(img)
clean_text=extracted.strip()
print(clean_text)

user_input = input("Enter your signature ")
if user_input.strip().lower() == clean_text.lower() :
    print("Access")
else:
    print("denied")
```
